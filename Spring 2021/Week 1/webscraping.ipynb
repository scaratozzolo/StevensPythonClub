```python
import requests
import pandas as pd
from datetime import datetime
```

```python
r = requests.get("https://www.basketball-reference.com/leagues/NBA_2021_games-january.html")
r
```

```python
t = pd.read_html(r.text, match="Home/Neutral")[0]
t["datestr"] = t[["Date", "Start (ET)"]].apply(lambda x: f"{x[0]} {x[1]}", axis=1)
t['date2'] = pd.to_datetime(t['datestr'])
t[t.date2 > datetime(2021, 1, 15)]
```
