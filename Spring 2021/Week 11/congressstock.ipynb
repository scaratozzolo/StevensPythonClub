```python
sub = df.loc[(df.index >="2021") & (df["ticker"] != "--") & (df["type"] == "Purchase")]
d = {i: [k,v] for i, (k, v) in enumerate(zip(sub.index, sub.ticker))}
d
```

```python
ticker = "DIS"
sub = df.loc[(df.index >="2020") & (df["ticker"] != "--") & (df["type"] == "Purchase") & (df['ticker'] == ticker)]
sub
```

```python
stock = yf.download(ticker, start=f"{sub.index.min().year}-01-01")["Adj Close"]
stock
```

```python
plt.plot(stock)
for k in sub.index:
    plt.vlines(k, ymin=stock.min(), ymax=stock.max(), color="red")
```
